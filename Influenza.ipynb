# Creating a graph database for the Legislative Branch.

```python
import yaml

with open("/Users/sahuguet/Documents/Dev/Influenza/legislators-current.yaml", 'r') as stream:
    try:
        data = yaml.load(stream)
    except yaml.YAMLError as exc:
        print(exc)
```

CREATE (:`Government Organization` { name:"US Senate"})
CREATE (:`Government Organization` { name:"House of Representatives"})

```python
for people in data:
    name = people['name']['official_full'].replace('"','\\"')
    role = people['terms'][-1]['type']
    party = people['terms'][-1]['party']
    if role == 'sen':
        print """CREATE (:Person:Senator { name: "%s", party: "%s" })""" % (name, party)
    elif role == 'rep':
        print """CREATE (:Person:Representative { name: "%s", party: "%s" })""" % (name, party)
```

-- Senators are MEMBER_OF the US Senate.
MATCH (n:Senator)
MATCH (senate:`Government Organization` { name: "US Senate"})
CREATE (n)-[:MEMBER_OF]->(senate);

-- Representatives are MEMBER_OF the House.
MATCH (n:Representative)
MATCH (house:`Government Organization` { name: "House of Representatives"})
CREATE (n)-[:MEMBER_OF]->(house);
